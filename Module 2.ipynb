```python
#2.6
df1 = pd.read_csv('C:\\Users\\Sunny\\Downloads\\Large_Cap\\Large_Cap\\ADANIPORTS.csv')
df2 = pd.read_csv('C:\\Users\\Sunny\\Downloads\\Large_Cap\\Large_Cap\\ASIANPAINT.csv')
df3 = pd.read_csv('C:\\Users\\Sunny\\Downloads\\Large_Cap\\Large_Cap\\AXISBANK.csv')
df4 = pd.read_csv('C:\\Users\\Sunny\\Downloads\\Large_Cap\\Large_Cap\\BAJFINANCE.csv')
df5 = pd.read_csv('C:\\Users\\Sunny\\Downloads\\Large_Cap\\Large_Cap\\HDFC.csv')
```

```python
df1 = df1[df1.Series == 'EQ']
df2 = df2[df2.Series == 'EQ']
df3 = df3[df3.Series == 'EQ']
df4 = df4[df4.Series == 'EQ']
df5 = df5[df5.Series == 'EQ']
```

```python
df6 = pd.concat([df1['Close Price'], df2['Close Price'],df3['Close Price'], df4['Close Price'], df5['Close Price']], axis=1)
df6.columns = ['Adani_Ports', 'Asian_Paint', 'Axis_Bank', 'Baj_Finance','HDFC']
df6.head()
```

```python
df6.dropna(inplace=True)
df6.head()
```

```python
df6['Adani_Port_Per_Change'] = df6['Adani_Ports'].pct_change()*100
df6['Asian_Paint_Per_Change'] = df6['Asian_Paint'].pct_change()*100
df6['Axis_Bank_Per_Change'] = df6['Axis_Bank'].pct_change()*100
df6['Baj_Finance_Per_Change'] = df6['Baj_Finance'].pct_change()*100
df6['HDFC_Per_Change'] = df6['HDFC'].pct_change()*100
df6.head()
```

```python
df6.dropna(inplace=True)
df6.head()
```

```python
df7 = df6.iloc[:,5:]
df7.head()
```

```python
df8 = pd.read_csv('C:\\Users\\Sunny\\Downloads\\Nifty50\\Nifty50\\Nifty50.csv')
df8
```

```python
df8['Per_change'] = df8.Close.pct_change()*100
df8.dropna(inplace=True)
df8.head()
```

```python
df8.Per_change.rolling(7).mean()
```

```python
#2.8
plt.figure(figsize=(15,5))
plt.plot(df7.Adani_Port_Per_Change.rolling(7).mean(), label='Adani_Port')
plt.plot(df8.Per_change.rolling(7).mean(), label= 'Nifty50')
plt.legend()
```

```python
#2.9
plt.figure(figsize=(15,5))
plt.plot(df8.Per_change.rolling(21).mean(),label='21day')
plt.plot(df8.Per_change.rolling(34).mean(),label='34day')
plt.legend()
```

```python
#2.10
Rolling_14 = df8.Per_change.rolling(14).mean()
BU = Rolling_14 + 2*df8.Per_change.rolling(14).std()
BL = Rolling_14 - 2*df8.Per_change.rolling(14).std()
plt.figure(figsize=(15,5))
plt.plot(Rolling_14, label='MA')
plt.plot(BU, label='BU')
plt.plot(BL,label='BL')
plt.legend()
```
