```python
#1.6
import numpy as np
df['Day_Perc_Change'] = df['Close Price'].pct_change()*100
df.head()
```

```python
df.fillna(0,inplace=True)
df.head()
```

```python
#1.7
def trend(x):
    if (x>-0.5) & (x<0.5):
        return('Slight or No Change')
    elif (x>0.5) & (x<1):
        return('Slight Positive')
    elif (x<-0.5) & (x>-1):
        return('Slight Negative')
    elif (x>1) & (x<3):
        return('Positive')
    elif (x<-1) & (x>-3):
        return('Negative')
    elif (x>3) & (x<7):
        return('Among Top Gainers')
    elif (x<-3) & (x>-7):
        return('Among Top Losers')
    elif (x>7):
        return('Bull Run')
    elif (x<-7):
        return('Bull Drop')
    else:
        pass
```

```python
df['Trend'] = df['Day_Perc_Change'].apply(trend)
df.head()
```

```python
#1.8
df.groupby('Trend')['Total Traded Quantity'].agg(['mean', 'median'])
```
